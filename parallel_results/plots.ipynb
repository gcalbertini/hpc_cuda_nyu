```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
batch_sizes = [8, 16, 32, 64, 128, 256]
serial_times_50k = [6.64229,  6.59965, 6.59217,  6.77139,  6.66895,  6.66124]
parallel_times_50k = [0.360022, 0.7005051, 1.348442, 2.685624, 4.36974, 4.6987]
speedups_50k = np.divide(serial_times_50k, parallel_times_50k)
test_errors_50k = [1.87238, 2.2211, 5.29928, 3.90475, 17.2356, 8.31485]

fig, ax = plt.subplots()
plt.scatter(batch_sizes, speedups_50k)
plt.plot(batch_sizes, speedups_50k, '--')
ax.set_xticks([8, 16, 32, 64, 128, 256])
ax.set_yticks(np.linspace(min(speedups_50k), max(speedups_50k), num=15))
plt.rc('xtick', labelsize=9)  # fontsize of the x tick labels
fig.suptitle('Speedup for Batch Size using N = 50000', fontweight="bold")
plt.grid(visible=True, which='major', axis='both')
plt.ylabel("Speedup")
plt.show()
plt.savefig('../generated_data/speedup.png')
```

```python
fig, ax = plt.subplots()
plt.scatter(batch_sizes, test_errors_50k)
plt.plot(batch_sizes, test_errors_50k, '--', label='Parallel')
fig.suptitle('Test Loss per Batch Size using N = 50000', fontweight="bold")
ax.set_xticks([8, 16, 32, 64, 128, 256])
plt.grid(visible=True, which='major', axis='both')
plt.ylabel("Test Loss")
plt.legend()
plt.show()
plt.savefig('../generated_data/test_batch_loss_50k.png')
```

```python
batch_errors_cuda_10k = [8.64418, 10.4194, 18.0644, 19.4555, 28.0262, 31.1045]
batch_errors_cuda_50k = [1.57641, 2.11554, 5.2852, 4.08888, 16.7898, 8.17022]
batch_errors_cuda_100k = [1.46496, 1.56097, 2.28265, 3.80604, 3.19115, 76.6081]
fig, ax = plt.subplots()
plt.scatter(batch_sizes, batch_errors_cuda_10k)
plt.plot(batch_sizes, batch_errors_cuda_10k, '--', label='Parallel,  N = 10000')
plt.scatter(batch_sizes, batch_errors_cuda_50k)
plt.plot(batch_sizes, batch_errors_cuda_50k, '--', label='Parallel,  N = 50000')
plt.scatter(batch_sizes, batch_errors_cuda_100k)
plt.plot(batch_sizes, batch_errors_cuda_100k, '--', label='Parallel,  N = 100000')
fig.suptitle('Test Loss per Batch Size', fontweight="bold")
ax.set_xticks([8, 16, 32, 64, 128, 256])
plt.grid(visible=True, which='major', axis='both')
plt.ylabel("Test Loss")
plt.legend()
plt.show()
plt.savefig('../generated_data/test_batch_loss_cuda.png')
```
